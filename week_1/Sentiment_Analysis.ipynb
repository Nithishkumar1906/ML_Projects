Training using Dataserts

```python
import numpy as np
import pandas as pd
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
dataset = pd.read_csv(r"C:\Users\hp\Desktop\Tech Mango\Datasets\Book_reviews.csv")
```

```python
dataset.head()
```

```python
dataset['review title'][:5]
```

```python
dataset['Reviews'] = dataset['review title'].apply(lambda X : re.sub(r"([^\w\s])","",X.lower())) 
```

```python
dataset.head(10)
```

```python
vader_sentiment = SentimentIntensityAnalyzer()
```

```python
dataset['Sentiment_score'] = dataset['Reviews'].apply(lambda X : vader_sentiment.polarity_scores(X)['compound'])
```

```python
dataset.head()
```

```python
bins = [-1,-0.1,0.1,1]
label = ["Positive","Neutral","Negative"]
dataset['Sentiment_Label'] = pd.cut(dataset['Sentiment_score'],bins,labels=label)
```

```python
dataset.head()
```

```python
dataset['Sentiment_Label'].value_counts().plot.bar()
```
