```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

print("Libraries imported successfully!")

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

print(f"\nOriginal dataset loaded.")
print(f"Original number of training images: {len(x_train)}")
print(f"Original number of testing images: {len(x_test)}")
print("There are 10 classes of clothing items.")


class_a = 0
class_b = 1

# Filter training data
train_mask = np.logical_or(y_train == class_a, y_train == class_b)
x_train_small = x_train[train_mask]
y_train_small = y_train[train_mask]

# Filter test data
test_mask = np.logical_or(y_test == class_a, y_test == class_b)
x_test_small = x_test[test_mask]
y_test_small = y_test[test_mask]

print(f"\nSmall binary dataset created.")
print(f"New number of training images: {len(x_train_small)}")
print(f"New number of testing images: {len(x_test_small)}")


x_train_small = x_train_small.astype('float32') / 255.0
x_test_small = x_test_small.astype('float32') / 255.0

# Reshape images for the CNN model
x_train_small = np.expand_dims(x_train_small, -1)
x_test_small = np.expand_dims(x_test_small, -1)

print("\nData preprocessing complete.")
print(f"Training data shape: {x_train_small.shape}")


class_names = ['T-shirt/top', 'Trouser']

plt.figure(figsize=(10, 5))
print("\nDisplaying 10 sample images from your new practice dataset:")
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train_small[i].reshape(28, 28), cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train_small[i]])
plt.show()


print("x_train_small  <-- Use this for training images")
print("y_train_small  <-- Use this for training labels")
print("x_test_small   <-- Use this for testing images")
print("y_test_small   <-- Use this for testing labels")
```

```python

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

print(" Step 1: Libraries imported successfully!")

(x_train_full, y_train_full), (x_test_full, y_test_full) = keras.datasets.fashion_mnist.load_data()


train_mask = np.logical_or(y_train_full == 0, y_train_full == 1)
test_mask = np.logical_or(y_test_full == 0, y_test_full == 1)

x_train, y_train = x_train_full[train_mask], y_train_full[train_mask]
x_test, y_test = x_test_full[test_mask], y_test_full[test_mask]

x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

print("\n Step 2: Dataset loaded and prepared.")
print(f"   - Training data shape: {x_train.shape}")
print(f"   - Testing data shape:  {x_test.shape}")

# Step 3: Build the CNN Model Architecture
model = keras.Sequential([
    # Input Layer: A convolutional layer with 32 filters
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),

    # Second convolutional layer with 64 filters
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Flatten the 2D image data into a 1D vector
    layers.Flatten(),

    # A dense layer for classification
    layers.Dense(64, activation='relu'),

    # Output Layer: A single neuron with a sigmoid activation for binary (0 or 1) output
    layers.Dense(1, activation='sigmoid')
])

print("\n Step 3: CNN model built successfully.")
model.summary()

# Step 4: Compile the Model

model.compile(
    optimizer='adam',
    loss='binary_crossentropy', # Use this loss function for two-class problems
    metrics=['accuracy']
)

print("\n Step 4: Model compiled successfully.")

# Train the Model
print("\n Step 5: Starting model training...")
history = model.fit(
    x_train,
    y_train,
    epochs=10,
    validation_data=(x_test, y_test)
)
print("Training complete.")
print("\n Step 6: Evaluating model performance...")
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"\n Final Test Accuracy: {test_acc*100:.2f}%")

# Plot training & validation accuracy values
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

plt.tight_layout()
plt.show()
```
